```python
def data_200(log_addr):
    X, labels, ranking, steps_q, steps_c,wave_q= load_data(log_addr)
    X_new=[]
    labels_new=[]
    ranking_new=[]
    steps_q_new=[]
    steps_c_new=[]
    wave_q_new=[]
    for i in range(5000):
        diff=steps_c[i][0]-steps_q[i][0]
        if abs(diff)<200:
            X_new.append(X[i])
            labels_new.append(labels[i])
            ranking_new.append(ranking[i])
            steps_q_new.append(steps_q[i])
            steps_c_new.append(steps_c[i])
            wave_q_new.append(wave_q[i])
    X_new= np.array(X_new)
    ranking_new = np.array(ranking_new)
    labels_new = np.array(labels_new)
    return X_new, labels_new, ranking_new,steps_q_new,steps_c_new,wave_q_new
```

```python


# Build NN
dropout_rate=0.2
inputs = tf.keras.Input(shape=(n, n, 1))
#x = tf.keras.layers.Conv2D(n, (2,2),padding = 'same', activation="relu")(inputs)
#x = tf.keras.layers.Conv2D(n, (2,2),padding = 'same', activation="relu")(x)


x = inputs
 
z=tf.keras.layers.Flatten()(x)
z = tf.keras.layers.Dense(num_classes, activation = 'sigmoid')(z)
#z = tf.keras.layers.Dense(num_classes, activation = 'sigmoid')(z)
z = tf.keras.layers.Dense(num_classes, activation = 'sigmoid')(z)

#z = tf.keras.layers.Dropout(dropout_rate)(z)
out = tf.keras.layers.Dense(num_classes, activation = 'sigmoid')(z)
model = tf.keras.models.Model(inputs, out)
opt = tf.keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-8)
model.compile(optimizer = opt,
          loss = tf.keras.losses.CategoricalCrossentropy(from_logits=False),
          metrics = ['accuracy'])
print(model.summary())


# Train NN
epochs = 200
history = model.fit(X_train, y_train, batch_size=batch_size, validation_data = (X_test, y_test), validation_freq = 1, epochs=epochs, verbose=2, shuffle = True)
```
