```python
import os, inspect  # for current directory

import numpy as np
import matplotlib.pyplot as plt

```

```python
a=6
SMALL_SIZE = 8+a
MEDIUM_SIZE = 10+a
BIGGER_SIZE = 12+a
plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)
prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']
```

```python
n = 20 # 20, 25
arch = 3
possible_classes = ['c_q', 'c_T', 'q_T']
classes = possible_classes[0]

for n in [15, 20, 25]:
    filename = os.path.join('results', 'n_' + str(n) , str(arch) + classes, 'num_nodes_'+ str(n) + '_nettype_' + str(arch) + classes + '_average_'+str(n)+'.npz')

    file = np.load(filename)
    # train_loss, test_accuracy, train_accuracy, test_loss
    train_loss = file['arr_0']
    test_accuracy = file['arr_1']
    train_accuracy = file['arr_2']
    test_loss = file['arr_3']

    plt.ylim(0.0, 1.0)
    stop = 200
    if len(test_accuracy) > stop:
        plt.plot(np.linspace(0.0, len(test_accuracy)-stop, len(test_accuracy)-stop), test_accuracy[0:stop],'-', label = 'n='+str(n), color=colors[0])
        plt.plot(np.linspace(0.0, len(test_loss) -stop, len(test_loss)-stop), test_loss[0:stop],'-.', color=colors[0])
    else:
        plt.plot(np.linspace(0.0, len(test_accuracy), len(test_accuracy)), test_accuracy,'-', label = 'n='+str(n))
    plt.xlabel('epochs')
    plt.ylabel('accuracy')
    plt.legend()
#plt.savefig('random.pdf', bbox_inches='tight')
```
